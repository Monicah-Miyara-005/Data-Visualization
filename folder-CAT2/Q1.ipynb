```python
import pandas as pd
import numpy as np

#Question 1 
data = {
    'ID': ['P001', 'P002', 'P003', 'P004', 'P005', 'P006'],
    'Pulse': [72, 85, 60, 78, 65, 90],
    'Systolic': [120, 135, 110, 125, 115, 140],
    'Diastolic': [80, 85, 70, 78, 75, 90],
    'Cholesterol': [190, 220, 180, 200, 210, 230],
    'AgeGroup': ['Adult', 'Adult', 'Senior', 'Adult', 'Senior', 'Adult']
}

df = pd.DataFrame(data)

#(a)  Pulse statstics 
mean_pulse = df['Pulse'].mean()
median_pulse = df['Pulse'].median()
std_pulse = df['Pulse'].std(ddof=0)
var_pulse = df['Pulse'].var()
min_pulse = df['Pulse'].min()
max_pulse = df['Pulse'].max()


print(f"Mean: {mean_pulse}")
print(f"Median: {median_pulse}")
print(f"Std deviation: {std_pulse}")
print(f"Variance: {var_pulse}")
print(f"Min: {min_pulse}")
print(f"Max: {max_pulse}")
```

```python
#(b) Covariance matrix and Eigen decomposition

numeric_df = df[['Pulse', 'Systolic', 'Diastolic', 'Cholesterol']]
cov_matrix = numeric_df.cov()
print("The covariance matrix:")
print(cov_matrix)

# Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print(f"Eigen Values: {eigenvalues}")
print("Eigen Vectors:")
print(eigenvectors)
```

```python
#(c) Cholesterol variability 
chol_var = df['Cholesterol'].var()

#Q1
q1 = df['Cholesterol'].quantile(0.25)
#Q3
q3 = df['Cholesterol'].quantile(0.75)
#Interqurtile range 
iqr = q3 - q1
print(f"Cholesterol variability : {chol_var}")
print(f"Q1:{q1}")
print(f"Q3:{q3}")
print(f"Interquartile range:{iqr}")
```

```python
#(d) Detect Pulse anomalies using Z-score
# Z-score = (x - mean) / std
df['Pulse_Zscore'] = (df['Pulse'] - mean_pulse) / std_pulse

print(df[['ID', 'Pulse', 'Pulse_Zscore']])
```
