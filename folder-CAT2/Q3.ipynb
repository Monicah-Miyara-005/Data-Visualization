```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

#Question 3 

data = {
    'ID': ['EN001', 'EN002', 'EN003', 'EN004', 'EN005', 'EN006'],
    'Temp': [24, 28, 22, 27, 23, 26],
    'Humidity': [60, 55, 70, 58, 72, 62],
    'EnergyUsage': [350, 400, 300, 420, 320, 380],
    'WindSpeed': [10, 12, 8, 11, 7, 9],
    'Location': ['Residential', 'Residential', 'Industrial', 'Residential', 'Industrial', 'Residential']
}

#(a)Engineer predictive features 

df = pd.DataFrame(data)
df['Temp_Squared'] = df['Temp'] ** 2
df['Temp_Humidity_Interaction'] = df['Temp'] * df['Humidity']
df['Energy_Per_Temp'] = df['EnergyUsage'] / df['Temp']
df['Log_Energy'] = np.log(df['EnergyUsage'])
df['Z_Temp'] = (df['Temp'] - df['Temp'].mean()) / df['Temp'].std()
df['Z_Humidity'] = (df['Humidity'] - df['Humidity'].mean()) / df['Humidity'].std()


print("Engineered Features:")
print(df[['ID', 'Temp', 'Temp_Squared', 'Temp_Humidity_Interaction','Energy_Per_Temp', 'Z_Temp','Log_Energy']])

```

```python

#(b) grouped statistics across locations

residential = df[df['Location'] == 'Residential']
industrial = df[df['Location'] == 'Industrial']

print("Residential Statistics:")
print(f"  Energy - Mean: {residential['EnergyUsage'].mean():.2f}, Std: {residential['EnergyUsage'].std():.2f} ,Max:{residential['EnergyUsage'].max()}, Min:{residential['EnergyUsage'].min()}")
print(f"  Temp - Mean: {residential['Temp'].mean():.2f}")
print(f"  Humidity - Mean: {residential['Humidity'].mean():.2f}")
print(f"Wind speed - Mean:{residential['WindSpeed'].mean():.2f}")


print("\nIndustrial Statistics:")
print(f"  Energy - Mean: {industrial['EnergyUsage'].mean():.2f}, Std: {industrial['EnergyUsage'].std():.2f} ,Max:{industrial['EnergyUsage'].max()}, Min:{industrial['EnergyUsage'].min()}")
print(f"  Temp - Mean: {industrial['Temp'].mean():.2f}")
print(f"  Humidity - Mean: {industrial['Humidity'].mean():.2f}")
print(f"Wind speed - Mean:{industrial['WindSpeed'].mean():.2f}")


```

```python

#(c) Construct regression model for EnergyUsage
# Independent variable (X) is Temp,Humidity,Windspeed and dependent variable (y) is Energy usage
X = df[['Temp', 'Humidity', 'WindSpeed']]
y = df['EnergyUsage']

# Add a constant (intercept) to the model
X = sm.add_constant(X)

# Fit the Ordinary Least Squares (OLS) model
model = sm.OLS(y, X).fit()

print("Regression Analysis")
print(f"Intercept : {model.params['const']:.2f}")
print(f"Temperature Coefficient : {model.params['Temp']:.2f}")
print(f"R-squared value: {model.rsquared:.4f}")
```
