```python
import pandas as pd
import numpy as np

#Question 2 

data = {
    'ID': ['E001', 'E002', 'E003', 'E004', 'E005', 'E006'],
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'UnitsSold': [200, 230, 210, 250, 220, 260],
    'Promotion': ['No', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'Revenue': [4000, 4600, 4200, 5000, 4400, 5200],
    'Customers': [120, 140, 130, 150, 135, 160]
}

df = pd.DataFrame(data)

#(a) revenue growth rates 
df['GrowthRate'] = df['Revenue'].pct_change()


print(df[['Month', 'Revenue', 'GrowthRate']])
```

```python

#(b) dispersion measures for Customers and Revenue 
customer_range = df['Customers'].max() - df['Customers'].min()
customer_std = df['Customers'].std()
customer_var = df['Customers'].var()

revenue_range = df['Revenue'].max() - df['Revenue'].min()
revenue_std = df['Revenue'].std()
revenue_var = df['Revenue'].var()

print("Dispersion Measures for customers")
print(f"Range: {customer_range}")
print(f"Standard Deviation: {customer_std}")
print(f"Variance: {customer_var}")

print("Dispersion Measures for revenue")
print(f"Range: {revenue_range}")
print(f"Standard Deviation: {revenue_std}")
print(f"Variance: {revenue_var}")
```

```python
#(c) Pearson correlation between UnitsSold and Revenue 
pearson_corr = df['UnitsSold'].corr(df['Revenue'])
print(f"Pearson Correlation: {pearson_corr}")
```
